```python
from sqlalchemy import create_engine
import pandas as pd
```

Load dataset

```python
df = pd.read_csv("ev_charging_data.csv", parse_dates=["datetime"])
```

```python
engine = create_engine(
    "postgresql+psycopg2://postgres:Nikeade1@localhost:5432/ev_pricing"
)

# Test connection
with engine.connect() as conn:
    print("Connected successfully!")
```

```python
df.to_sql(
    "charging_sessions",
    engine,
    if_exists="append",
    index=False
)
```

Load static_pricing data

```python
static_df = pd.read_csv("static_pricing_data.csv")
```

```python
static_df.to_sql(
    "static_pricing",
    engine,
    if_exists="replace",
    index=False
)
```
